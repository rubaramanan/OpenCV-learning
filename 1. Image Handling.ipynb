```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
from PIL import Image # pillow library
from IPython.display import display
import ipywidgets
```

### Read the image

```python
# read image
im = cv2.imread('cat.jpg')

cv2.imshow('sample image', im)

cv2.waitKey(0) # hold the image window until user closs
cv2.destroyAllWindows() # remove created ui window
```

```python
# show image using matplot lib

plt.imshow(im)
# plt.waitforbuttonpress()
# plt.close('all')
```

```python
# read image with different color code
# -1, cv2.IMREAD_COLOR - Load a color image any transparency
# 0, cv2.IMREAD_GRAYSCALE - load image in grayscale mode
# 1, cv2.IMREAD_UNCHANGED - loads image as such including alpha channel

im = cv2.imread('cat.jpg', 1)

cv2.imshow('sample image', im) # window name, image variable

cv2.waitKey(0) # hold the image window until user close
cv2.destroyAllWindows() 
```

```python
plt.imshow(im)
```

```python
# read the image in different color code
im = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('image in grayscale', im)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# we have to make array between 0-255
randomByteArray = bytearray(os.urandom(120000))
flattenArray = np.array(randomByteArray)
flattenArray
```

### Write image

```python
# convert the array to 300x400 matrix
grayImage = flattenArray.reshape((300, 400))

cv2.imwrite('gray_image.jpg', grayImage)

grayImage3 = cv2.imread('gray_image.jpg')
cv2.imshow('gray_image', grayImage3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# show image in output using pillow and display library
display(Image.fromarray(grayImage3))
```

```python
# color rgb image from array
rGBImage = flattenArray.reshape(100, 400, 3)
cv2.imwrite('rgb_image.jpg', rGBImage)

rGBImage3 = cv2.imread('rgb_image.jpg')
display(Image.fromarray(rGBImage3))
```

```python
# display the r,g,b channels of the image
blue_chan, green_chan, red_chan = cv2.split(rGBImage)
blue_image = cv2.merge([blue_chan, blue_chan, blue_chan])
green_image = cv2.merge([green_chan, green_chan, green_chan])
red_image = cv2.merge([red_chan, red_chan, red_chan])
cv2.imshow('blue image', blue_image)
cv2.imshow('red image', red_image)
cv2.imshow('greed image', green_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### convert color sclae of image

#### Using ipywidgets

Finally, we can use an ipywidget to display the image. You will need to convert the image from an array to bytes.

ipywidgets are great to use for displaying images and videos remotely within a Jupyter Notebook.

In the code below, we are displaying the original image of the cat that we read in using the CV2 in the previous code block.

```python
#image encoding
encode = cv2.imencode('.jpg', im4)[1]
bytes_from_array = bytes(encode)

image_widgets = ipywidgets.Image(format='jpeg')
image_widgets.value = bytes_from_array

display(image_widgets)
```

#### Image resize

```python
im = cv2.imread('cat.jpg', 1)
re_im = cv2.resize(im, (200,200))
cv2.imshow('Resize image', re_im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# resize the image by half of image pixel
re_im = cv2.resize(im, (0, 0), fx=0.5, fy=0.5)
cv2.imshow('Resize image', re_im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Rotating image

```python
ro_image = cv2.rotate(re_im, cv2.ROTATE_90_CLOCKWISE)
cv2.imshow('Rotate image', ro_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
